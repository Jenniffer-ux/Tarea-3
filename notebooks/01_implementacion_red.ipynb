```python
import sys
sys.path.append('/content/drive/MyDrive/tarea-redes-neuronales-final/src')

from neural_network import NeuralNetwork
from data_preprocessing import cargar_y_preprocesar

import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error

X_train, X_test, y_train, y_test = cargar_y_preprocesar(
    path="/content/drive/MyDrive/tarea-redes-neuronales-final/data/synthetic_data.csv"
)

modelo = NeuralNetwork(layers=[X_train.shape[1], 16, 8, 1], activation='tanh')
modelo.train(X_train, y_train, epochs=1000, learning_rate=0.001)
y_pred, _ = modelo.forward(X_test)
y_pred = y_pred[-1]

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(f"MSE: {mse:.4f}")
print(f"MAE: {mae:.4f}")

plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel("Real")
plt.ylabel("Predicho")
plt.title("Demanda: Real vs Predicho")
plt.grid(True)
plt.savefig("/content/drive/MyDrive/tarea-redes-neuronales-final/results/demanda_prediccion.png")
plt.show()
```
